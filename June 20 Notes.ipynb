```python
count = 0
lines = []

fh = open('Pride and Prejudice.txt', 'r', encoding = 'utf-8')
for line in fh:
    lines.append(line)
    count += 1
print(len(lines))
print(count)
fh.close()
```

```python
words = []
for line in lines:
    lwords = line.split()
    for word in lwords:
        words.append(word)
for i in range(20):
    print(words[i])
```
